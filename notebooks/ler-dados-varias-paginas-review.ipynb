```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = 'https://www.amazon.com.br/Hobbit-p%C3%B4ster-J-R-R-Tolkien/product-reviews/8595084742/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews&pageNumber='

for pageNumber in range(1, 260):
    page = requests.get(url + str(pageNumber))
    soup = BeautifulSoup(page.text, 'html.parser')
    reviews = soup.select('.review')

    if not reviews:
        break;
        
    print('\n**************** PÁGINA ' + str(pageNumber) + ' *********************************\n')
    
    for review in reviews:
        reviewTextContent = review.select('.review-text-content')
        quantidadeElementosSpan = len(reviewTextContent[0].select('span'))
        texto = reviewTextContent[0].select('span')[quantidadeElementosSpan - 1]
        print(texto)
        print('\n===================================================\n')
    
```
